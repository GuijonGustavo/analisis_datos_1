```python
import os
import sys
import numpy as np
import pandas as pd

sys.path.insert(0,os.path.join('..', '..'))
from procesamientos.scripts.func_analisis import *
from procesamientos.scripts.func_transformacion import *
from procesamientos.scripts.func_salida import *
```

```python
# carga de bases de datos
nombre_archivo='enasama_W18_0429_0505_ori_ver_anali_prelimi_estrato_reg.xlsx'
db=pd.read_excel(os.path.join('..','..','datos','limpios',nombre_archivo), engine='openpyxl')
```

```python
# selección de cuestionarios con folios válidos y que dieron autorización
b=db[(db['folio_val_dai']) & (db['cumple_autorizacion']=='SI')]

# asegurarse de que la clave de entidad sea un str
b['clave_entidad_dai']= b['clave_entidad_dai'].astype(str).str.zfill(2)

# cambio de variables con valores Y y N por 1 y 2 
b.replace({'Y':1, 'N':2}, inplace= True)

# creación de variable con nombre sexo
b.insert(0,'sexo',b['ds1'].replace({'F':'Mujeres','M':'Hombres'}))

#creación de variable llamada 'entidad'
b.insert(0,'entidad',b['nom_ent_largo'])

#creación de variable llamada 'region'
b.insert(0,'region',b['nom_region']) 
```

```python
# Mapeamos los valores
nivel_educativo_map = {
    1: 'Primaria incompleta (1 a 5 años)',
    2: 'Primaria completa (6 años)',
    3: 'Secundaria incompleta (1 a 2 años)',
    4: 'Secundaria completa o equivalente (3 años)',
    5: 'Bachillerato incompleto (1 a 2 años)',
    6: 'Bachillerato completo o equivalente (aprox. 3 años)',
    7: 'Estudios Universitarios incompletos (1 a 3 años)',
    8: 'Estudios Universitarios completos (4 a 5 años)',
    9: 'Estudios de Posgrado (2 a 4 años)',
    99: 'No sabe/No contesta'
}
```

### Nacional

```python
# Asignamos el nivel educativo a cada registro usando un mapeo
b['nivel_educativo'] = b['ds10'].map(nivel_educativo_map)

# Agrupamos los datos por sexo, grupo etario y nivel educativo, y contamos el número de registros en cada grupo
al_nac_edu = b.groupby(['sexo', 'grupo_etario', 'nivel_educativo']).size().reset_index(name='conteo')

# Calculamos el porcentaje que representa cada grupo sobre el total de registros
al_nac_edu['porcentaje'] = (al_nac_edu['conteo'] / al_nac_edu['conteo'].sum()) * 100

# Mostramos las primeras filas del resultado
al_nac_edu.head
```

### Estatal

```python
# Asignamos el nivel educativo a cada registro usando un mapeo
#b['nivel_educativo'] = b['ds10'].map(nivel_educativo_map)

# Agrupamos los datos por clave de entidad, nombre de la entidad, sexo y nivel educativo, y contamos el número de registros en cada grupo
# al_edo_edu = b.groupby(['clave_entidad_dai', 'grupo_etario', 'entidad', 'sexo', 'nivel_educativo']).size().reset_index(name='conteo')
al_edo_edu = b.groupby(['grupo_etario', 'entidad', 'sexo', 'nivel_educativo']).size().reset_index(name='conteo')

# Calculamos el porcentaje de cada nivel educativo dentro de cada entidad y sexo
# Para cada entidad y sexo, dividimos el conteo de cada nivel educativo por el total de registros de esa combinación
al_edo_edu['porcentaje'] = al_edo_edu.groupby(['entidad', 'sexo', 'grupo_etario'])['conteo'].transform(lambda x: (x / x.sum()) * 100)

# Mostramos las primeras dos filas del resultado
al_edo_edu.head
```

### Regional

```python
# Agrupamos los datos por región, sexo y nivel educativo, y contamos el número de registros en cada grupo
al_reg_edu = b.groupby(['region', 'sexo', 'grupo_etario', 'nivel_educativo']).size().reset_index(name='conteo')

# Calculamos el porcentaje de cada nivel educativo dentro de cada región y sexo
# Para cada combinación de región y sexo, dividimos el conteo de cada nivel educativo por el total de registros de esa combinación
al_reg_edu['porcentaje'] = al_reg_edu.groupby(['region', 'sexo', 'grupo_etario'])['conteo'].transform(lambda x: (x / x.sum()) * 100)

# Mostramos las primeras dos filas del resultado
al_reg_edu.head
```

```python
# Validamos a nivel nacional

# Calculamos el nivel educativo
b['nivel_educativo'] = b['ds10'].map(nivel_educativo_map)

# Agrupamos y contamos
al_nac_edu = b.groupby(['sexo', 'grupo_etario', 'nivel_educativo']).size().reset_index(name='conteo')

# Calculamos los porcentajes
al_nac_edu['porcentaje'] = (al_nac_edu['conteo'] / al_nac_edu['conteo'].sum()) * 100

# Validación de resultados
def validar_resultados(df):
    # Validar conteos por grupo
    print("\nValidación de conteos por grupo:")
    conteo_por_grupo = df.groupby(['sexo', 'grupo_etario'])['conteo'].sum()
    print(conteo_por_grupo)
    
    # Validar que los porcentajes suman 100% por grupo
    print("\nValidación de porcentajes:")
    porcentajes_por_grupo = df.groupby(['sexo', 'grupo_etario'])['porcentaje'].sum()
    print(porcentajes_por_grupo)
    
    # Verificar que el total de conteos coincida con la suma de conteos por grupo
    total_conteos = df['conteo'].sum()
    print("\nTotal de conteos:", total_conteos)
    print("Conteos por grupo sumados:")
    conteos_por_grupo_sumados = df.groupby(['sexo', 'grupo_etario'])['conteo'].sum().sum()
    print(conteos_por_grupo_sumados)
    
    # Verificar que los porcentajes suman 100%
    print("\nSuma de porcentajes:", df['porcentaje'].sum())

# Ejecutar la validación
validar_resultados(al_nac_edu)
```
